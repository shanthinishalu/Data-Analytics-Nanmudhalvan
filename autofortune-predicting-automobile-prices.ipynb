```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from numpy import nan
```

```python
automobile_data = pd.read_csv('/kaggle/input/automobile-dataset/Automobile_data.csv')
automobile_data.head()
```

```python
automobile_data.shape
```

```python
automobile_data['normalized-losses'].value_counts()
```

```python
automobile_data['normalized-losses'].replace(['?'], [nan], inplace=True)
```

```python
automobile_data.isnull().sum()
```

```python
automobile_data.dropna(inplace=True)
```

```python
automobile_data.shape
```

```python
automobile_data['bore'].value_counts()
```

```python
# in our case, we are not dealing with missing values here, we are exploring plots, so we can drop all "?" values.
automobile_data.replace(['?'], [nan], inplace=True)
automobile_data.dropna(inplace=True)
```

```python
automobile_data.shape
```

```python
automobile_data.to_csv('/kaggle/working//automobiles_data_processed.csv', index=False)
```

```python
plt.figure(figsize=(12, 8))

automobile_data['price'] = automobile_data['price'].astype(int)
automobile_data['horsepower'] = automobile_data['horsepower'].astype(int)
sns.histplot(automobile_data['price'], color='red', kde=True)
plt.title('Automobiles Data')
plt.show()
```

```python
automobile_data['price']
```

```python
price_data = automobile_data['price']

plt.figure(figsize=(12, 8))
sns.kdeplot(price_data, color='blue')
plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.kdeplot(automobile_data['price'],
           color='blue')
sns.rugplot(x=automobile_data['price'], color='red')
plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.rugplot(price_data.astype(int),
            height=0.5, color='blue', )

plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.kdeplot(price_data.astype(int),
             fill=True, color='blue')

plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='horsepower', y='price', data=automobile_data, s=120)

plt.title('Automobile Data')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='horsepower', y='price', hue='num-of-cylinders', data=automobile_data, s=120)

plt.title('Automobile Data')
plt.show()
```

```python
sns.regplot(x='horsepower', y='price', data=automobile_data)
```

```python
plt.figure(figsize=(12, 8))
sns.regplot(x='highway-mpg', y='price', data=automobile_data)

plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.jointplot(x='horsepower', y='price', data=automobile_data)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.jointplot(x='horsepower', y='price', kind='reg', data=automobile_data)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.set_theme(style='whitegrid')
sns.jointplot(x='horsepower', y='price', kind='kde', data=automobile_data)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.set_theme(style='whitegrid')
sns.jointplot(x='horsepower', y='price', kind='hex', data=automobile_data)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(6, 6))

sns.kdeplot(x=automobile_data['horsepower'], y=automobile_data['price'], ax=ax)
sns.rugplot(x=automobile_data['horsepower'], color='limegreen', ax=ax)
sns.rugplot(y=automobile_data['price'], color='red', ax=ax)

plt.title('Automobile Data')
plt.show()
```
